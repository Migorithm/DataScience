## 1. How much missing data are we working with?

Get a count and percentage of each variable in the dataset to answer this question.

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
#get number of nulls in each column
#1

def count_null(df):
    num_of_null = Row([*(df.filter(col(c).isNull()).count() for c in df.columns)])
    null_df = spark.createDataFrame(num_of_null,df.columns)
    
    return null_df

null_df = count_null(weather)
null_df.toPandas()
```

```python
#2

def null_value_calculator(df):
    null_columns_counts = []
    numRows = df.count()
    for c in df.columns:
        nullRows = df.where(col(c).isNull()).count()
        if(nullRows >0):
            temp= c,nullRows, (nullRows/numRows)*100
            null_columns_counts.append(temp)
    
    dataframe = spark.createDataFrame(null_columns_counts,["Col_name","num_null","percentage"])
    return dataframe.toPandas()

null_value_calculator(weather)
```

## 8. Fill in NaN values with averages for the tempm and tempi columns

*Note: you will first need to compute the averages for each column and then fill in with the corresponding value.*

```python
weather.filter(col("tempm").isNull()).count()
```

```python
weather.filter(col("tempi").isNull()).count()
```

```python
def fill_with_mean(df,include=set()):
    #gonna get a set from the second argumnet
    avgs = df.agg(*(avg(c).alias(c) for c in include))  #changing name is handy way to specify coloumns
    return df.fillna(avgs.first().asDict(),subset=[*include]) #take the first row and as it is 'Row'type, you can change its type to dict.

fill_with_mean(weather,["tempm","tempi"]).filter(col('tempm').isNull() & col('tempi').isNull()).count()
```
