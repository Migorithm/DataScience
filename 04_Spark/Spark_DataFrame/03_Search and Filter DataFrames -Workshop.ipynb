### 1. Select the Name and Position of each player in the dataframe

```python
fifa.select(["Name","Position"]).show(5,False)
```

### 1.1 Display the same results from above sorted by the players names

```python
fifa.select("Name","Position").orderBy(fifa.Name.asc()).toPandas()
```

### 2. Select only the players who belong to a club begining with FC

```python
#One way
fifa.select("Name","Club").filter(fifa.Club.startswith("FC")).toPandas()
```

```python
#Another way
fifa.select("Name","Club").where(fifa.Club.like("FC%")).show(5,False)
```

### 3. Who is the oldest player in the dataset and how old are they?

Display only the name and age of the oldest player.

```python
#One way
fifa.orderBy(fifa.Age.desc()).select("Name","Age").show(1)
```

```python
#another way
fifa.select("Name","Age").sort(desc("Age")).show(1)
```

### 4. Select only the following players from the dataframe:

 - L. Messi
 - Cristiano Ronaldo

```python
#one way
fifa.filter(fifa.Name.isin(["L. Messi","Cristiano Ronaldo"])).toPandas()
```

```python
fifa[fifa.Name.isin(["L. Messi","Cristiano Ronaldo"])].toPandas()
```

### 5. Can you select the first character from the Release Clause variable which indicates the currency used?

```python
#without substring
fifa.select("Release Clause").limit(1).toPandas().loc[0].values[0][0]
```

```python
#with substring1
fifa.select("Release Clause", fifa["Release Clause"].substr(1,1)).show(5)

#fifa["Relase Clause"]  is preferred form when a field has space. 
```

### 6. Can you select only the players who are over the age of 40?

```python
#one way
fifa.filter(fifa.Age>40).toPandas()
```

```python
fifa.filter("Age>40").limit(4).toPandas()
```
