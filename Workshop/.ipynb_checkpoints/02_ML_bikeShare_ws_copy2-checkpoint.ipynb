# Bike Sharing Demand Modeling

이번 수업에서는 **캐글의 <span style="color:blue">Bike Sharing Demand</span> 경진대회 자료를 다뤄봅니다.** <br>


주어진 정보는 크게 2가지로 분류해 볼수 있습니다.<br>
    * 시간, 날짜 관련 정보 :: datetime, season, holiday, workingday  <br>
    * 날씨 관련 정보 :: weather, temp, atemp, humidity, windspeed   <br>

이 정보들을 이용해서 자전거 대여 횟수를 예측하는 분석과 머신러닝을 이용한 모델을 설계, 학습하여<br>
미지의 Test 데이타(날짜와 날씨 정보)를 입력해서 최종적으로 사람들이 얼마나 자전거를 대여할지를 예측해봅니다.

<p></p>
주어진 데이터가 대체로 정교한 편이라서 데이타 전처리 과정이 상대적으로 간편합니다.<br>

해당 자료는 <u><span style="color:blue">[다음 사이트](https://www.analyticsvidhya.com/blog/2015/06/solution-kaggle-competition-bike-sharing-demand/)</span></u>를 참고로 작성되었습니다.

### 컬럼 설명

(데이터는 <u><span style="color:blue">[다음의 링크](https://www.kaggle.com/c/bike-sharing-demand)</span></u>에서 다운받으실 수 있습니다)

* datetime :: hourly date + timestamp
* season ::  1 = spring, 2 = summer, 3 = fall, 4 = winter (1,2,3,4 분기로 분석하는것이 더 정확)
* holiday :: 1 = 주말을 제외한 국경일 등의 휴일, 0 = 휴일이 아닌 날
* workingday :: 1 = 주말 및 휴일이 아닌 주중, 0 = 주말 및 휴일
* weather  :: 
    * 1=맑음, 약간 구름 낀 흐림
    * 2=안개, 안개 + 흐림
    * 3=가벼운 눈, 가벼운 비 + 천둥
    * 4=심한 눈/비, 천둥/번개
* temp :: 온도(섭씨)
* atemp :: 체감온도(섭씨)
* humidity :: 상대습도
* casual :: 사전에 등록되지 않은 사용자가 대여한 횟수
* registered :: 사전에 등록된 사용자가 대여한 횟수
* count :: 대여 횟수

#### datetime을 세부적으로 나누어서 컬럼 추가하고 기존 컬럼 삭제하기

```python
train['datetime'] = pd.to_datetime(train['datetime'])

train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
# train["minute"] = train["datetime"].dt.minute
# train["second"] = train["datetime"].dt.second

print(train.shape)
train.drop('datetime', axis=1, inplace=True)
train.head(1)
```

```python
test['datetime'] = pd.to_datetime(test['datetime'])

test["year"] = test["datetime"].dt.year
test["month"] = test["datetime"].dt.month
test["day"] = test["datetime"].dt.day
test["hour"] = test["datetime"].dt.hour
# test["minute"] = test["datetime"].dt.minute
# test["second"] = test["datetime"].dt.second

print(test.shape)
test.drop('datetime', axis=1, inplace=True)
test.head(1)
```

```python
sub_train = train[["windspeed","season", "weather", "humidity", "day", "temp", "atemp"]] # ->두개로..

zero=sub_train[sub_train['windspeed']==0] # 예측을 해봄 -> 예측값을 넣을 것
non_zero=sub_train[sub_train['windspeed']!=0] # 예측모델을 만듬 -> train - test 분리

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(non_zero[["season", "weather", "humidity", "day", "temp", "atemp"]],
                                                  non_zero['windspeed'],test_size=0.2,random_state=4)
```

```python
X_train=X_train.apply(lambda x:round(x).astype('int64'),axis=1)
X_test=X_test.apply(lambda x:round(x).astype('int64'),axis=1)
y_train=y_train.astype('int64').map(lambda x: x- x%5)
y_test=y_test.astype('int64').map(lambda x: x- x%5)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
train_score=model.score(X_train,y_train)
test_score=model.score(X_test,y_test)

print("훈련:{:.3f}".format(train_score))
print("테스트:{:.3f}".format(test_score))
# train_score=sub_train.score(X_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)

# for i,j in zip(pred,y_test):
#     print(i,j)
print("훈련",model.score(X_train,y_train))
print("test",model.score(X_test,y_test))
```
