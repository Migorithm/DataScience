##### 2019년 의 영화 평점 데이터를 가져옵니다.

```python
import time

date = pd.date_range('20190101', periods=31, freq='D')

movie_date = []
movie_name = []
movie_point = []

for today in date:
    html = "http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=
                                   urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
    time.sleep(1)
```

```python
# 수집된 데이터 건수 확인
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 
                      'name':movie_name, 
                      'point':movie_point})
movie.head()
```

```python
# 인터넷의 문제로 미리 수집된 데이터를 이용합니다.
# 2019년 네이버 영화 평점 리스트(2019년 네이버 영화 평점 리스트.csv 파일이용)를 movie 변수로 불러옵니다.
movie = pd.read_csv('../data/2019년 네이버 영화 평점 리스트.csv')



movie.head()
```

```python
# date 컬럼을 datetime 타입으로 변경합니다.
movie['date'] = pd.to_datetime(movie['date'],format = "%Y-%m-%d")
movie.info()

#인포 확인해보니 point가 object로 되어있다  -> 변경



```

```python
# 평균 평점이 낮은 10개 영화를 시각화합니다.
movie.groupby(by='name').agg(['mean']).sort_values(by=[('point','mean')],ascending=True).head(10)
```

```python
# 영화별 평점 건수(리뷰 요일)를 계산합니다.

movie.drop(axis=1,columns='date').groupby(by='name').count().rename(columns={"point":"count"})
```

```python
# 평점 건수(리뷰 요일)가 높은 10개 영화를 시각화합니다.
movie.drop(axis=1,columns='date').groupby(by='name').count().rename(columns={"point":"count"}).sort_values(by='count',ascending=False).head(10)
```

```python
# 평점 건수(리뷰 요일)가 낮은 10개 영화를 시각화합니다.
movie.drop(axis=1,columns='date').groupby(by='name').count().rename(columns={"point":"count"}).sort_values(by='count',ascending=True).head(10)
```

```python
# 평균 평점이 높은 순서대로 영화 정보를 추출합니다.
#step1 평균평점?
tmp = movie.groupby(by='name').mean().sort_values(by='point',ascending=False)

tmp.index
```

```python
# 평균 평점 1위 영화의 평점 변화를 시각화합니다. plot() 사용
#step1 1위 영화의 이름은?
rank1= tmp.iloc[0].name

#step2 해당 영화를 바탕으로 새로운 테이블?
rank1_chart = movie[movie['name']==rank1]



rank1_chart.plot(x='date',figsize=(12,8))
plt.show()
```

```python
# 모든 영화에 대해 날짜별 평점 변화를 계산합니다.

movie_pivot = movie.pivot_table(index=['date'],columns='name',values='point')
movie_pivot.head()
```

```python
# 평점이 없는 경우 (na의 경우) 0으로 치환합니다.
movie_pivot.fillna(0,  inplace=True)
movie_pivot
```

```python
# date 컬럼에서 년, 월, 일, 요일 정보를 추출합니다.

movie['month'] = movie['date'].dt.month
movie['day']   = movie['date'].dt.day
movie['week']  = movie['date'].dt.week
movie['wday']  = movie['date'].dt.weekday

#movie['date'] 는 datetime을 datatype으로 갖는 시리즈. 시리즈의 메소드 중 .dt.day 는 데이트타임 타입의 요일 속성을 반환해준다.
movie

```

```python
# 영화의 요일별 평균 평점을 계산합니다.
movie[['name','wday','point']].groupby(by=['name','wday']).mean().rename(columns={'point':'mean'})
```

```python
# 인덱스를 재설정합니다. reset_index함수 이용
temp = movie[['name','wday','point']].groupby(by=['name','wday']).mean().rename(columns={'point':'mean'}).reset_index()
temp
```

```python
# "기생충" 영화의 요일별 평점을 시각화합니다...plot()사용

movie[movie['name']=='기생충'].groupby(by='wday').mean().drop(axis=1,columns=['month','day','week']).rename(columns={'point':'mean'}).plot()
```

```python
# 영화의 주별 평균 평점을 계산합니다.

movie[['name','week','point']].groupby(by=['name','week']).mean().rename(columns={'point':'mean'})
```

```python
temp = movie[['name','week','point']].groupby(by=['name','week']).mean().rename(columns={'point':'mean'}).reset_index()
temp
```

```python
#  평균 평점 10위 영화의 월별 평균 평점을 계산합니다.
index= movie.groupby(by=['name'])[['point']].mean().sort_values(by='point',ascending=False).head(10).index

movie[movie['name'].isin(index)].groupby(by=['name','month'])[['point']].mean().rename(columns={'point':'mean'})
```
