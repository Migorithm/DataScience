### Data loader - sklearn datasets.load_iris()



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
iris= datasets.load_iris()

print(iris.data) #values
print(iris.feature_names) #column names
print(iris.target)#answers
print(iris.target_names)


```

### Dataframe visualization

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)

#iris.target_names 은 스트링 밸류이고
#iris.target 은 0,1,2로 encoding 되어있으므로 decoding해주기 위해 iris.target의 번호를 인덱스번호로 인식 - > 인덱스에 해당하는 값을 적용시켜 
#새로운 np.array를 만들고 데이터프레임에 편입시킨다.

df['species'] = np.array([iris.target_names[i] for i in iris.target])
df
```

### 머신러닝 연결하기

iris Data Shuffle/Split

```python
from sklearn.model_selection import train_test_split

#note that connecting ML is nothing to do with DataFrame visualization

#to split the given data, we use train_test_split method
X_train,X_test, y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.2,random_state=0)
```

```python
#shape check

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Generating Model

```python
#import module
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(random_state=100) #seed for model
iris_tree.fit(X_train,y_train)#put train data and its answer
```

### Making prediction

```python
tree_pred = iris_tree.predict(X_test)
```

```python
#Check the result of prediction
#Compare that with real answer

print("prediction : \n", tree_pred)
print("answer : \n", y_test)

#You can do this to check the accuracy.. but if you have tons of millions of data, less likely. 
```

#### Accuracy evaluation
- score()
- accuracy_score()

```python
#1 rating the training result
print("{:.3f}".format(iris_tree.score(X_train,y_train)))

#2 rating the test result
print(iris_tree.score(X_test,y_test))

#성능을 평가하는데 predict가 없다?
#score()는 내부적으로 predict를 가동. 
```

```python
#3 different module - 매트릭스를 이용해 보겠다.

from sklearn.metrics import accuracy_score

accuracy_score(tree_pred, y_test)
```

### Graphvis visualization

```python
import graphviz
from sklearn.tree import export_graphviz

#graphvis를 구현하기 위한 모듈 sklearn.tree.export_graphviz

#훈련시킨 decision tree 모델을 넣는다. 
#out_file 로 익스포트한다. 
#class_names로 label 를 지정한다.
#feature_names로 feature 지정

#<optional> filled : 노드 색깔
#<optional> impurity : 불순도 체크 - 지니 계수로 나온다. default는 True

export_graphviz(iris_tree, out_file='prac1.dot', #해당 정보를 저장
                class_names=iris.target_names, #class_name = label
                  feature_names=iris.feature_names,
               )

with open('prac1.dot',encoding='utf-8') as f:
    dot_graph = f.read() #저장해놓은 정보를 한번에 불러내서 변수에 입력

dot = graphviz.Source(dot_graph) #변수에 입력된 소스로부터 graphviz 구현
display(dot)



#이미지 내보내기 (pdf파일이 default)
# -> dot.render(filename = "iris_tree")
```
