### Module Configuration


```python
import pandas as pd
from sklearn.svm import SVC
import numpy as np
#sklearn? scikit-learn- support vector machine    
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### Visualization with pandas

```python
df = pd.DataFrame(X,columns=iris.feature_names)
df['Sepecies'] = np.array([iris.target_names[i] for i in iris.target])
df
```

### Data Spliting and Shuffling


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.2,random_state=3)
```

### Model generation and Training, Testing

```python
svm = SVC(C=2)
svm.fit(X_train,y_train)

print(svm.score(X_train,y_train))
print(svm.score(X_test,y_test))
```
