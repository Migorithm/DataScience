```python
#이상치 처리
df['고급휘발유'] = np.where(df['고급휘발유']==0,np.nan,df['고급휘발유'])
df['휘발유'] = np.where(df['휘발유']==0,np.nan,df['휘발유'])
df['경유'] = np.where(df['경유']==0,np.nan,df['경유'])

df.isnull().sum()
#turns out, 108256 gas stations don't handle high-quality gasoline.


#실내 등유는 주로 가정에서 사용하는 팬히터, 스토브, 온풍기 등 가정 난방용 연료
#따라서 여름에 쓰일 일이 없다! 
```

```python
#널값 처리!  - 10만개 이상의 null값을 소거할수는 없다. 따라서 평균값으로 채운다.
df['고급휘발유'].fillna(df['고급휘발유'].mean(),inplace=True)
df['휘발유'].fillna(df['휘발유'].mean(),inplace=True)
df['경유'].fillna(df['경유'].mean(),inplace=True)
```

```python

df4 = df.drop(axis=1,columns=['년','월','일','실내등유']).set_index('기간')
#실내등유또한 위의 가정에서 구한 것과 같이 분석을 하는데 있어서 계절 특이성, 공급 특이성에 의한 방해요소로 작용하므로 제거해준다.

#유가 평균 구해보자! 
df4['유가평균'] = (df4['휘발유'] + df4['경유'] + df4['고급휘발유'])/3



#지역별 주유소 갯수를 구해보자
plt.figure(figsize=(20,14))
sns.countplot(x='구',data=df4)
```

### 데이터 프레임의 재조립이 필요 
- **피봇이나 그룹바이로는 numerical 데이터만 밸류로 인정되기때문에** 
- 우리가 관심있는 셀프여부나, 구에 대한 정보는 사라지게되거나 인덱스형태로만 존재한다.
- hue와 같은 키워드 매개변수에는 인덱스값이 올 수 없으므로, 해당 컬럼에 대한 값들을 유지한 새로운 데이터 프레임이 필요하다.
<br><br>
- 또한 concat, merge로도 이러한 부분이 해결되지 않는다. 
    - 중복된 값들이 하나로 합쳐지지 않기 때문. 



```python
#데이터 재조립!

lst = []
for i in df4['상호'].unique():
    lst_in_lst = []
    average_price = df4[df4['상호']==i]['유가평균'].mean() #유가평균의 연 평균
    ku = df4[df4['상호'] == i]['구'].unique()[0] 
    self = df4[df4['상호'] == i]['셀프여부'].unique()[0]
    brand = df4[df4['상호'] == i]['상표'].unique()[0]
    lst_in_lst.extend([ku,i,brand,self,average_price])
    lst.append(lst_in_lst)
lst

#df4[['구','상호','상표','셀프여부','유가평균']]
```

```python
df6 = pd.DataFrame(lst,columns=['구','상호','상표','셀프여부','유가평균'])
df6
```

**Q.** 유가가 유독 비쌌던 '종로구','중구','용산구'가 속한 서울 서북지역에는 SK에너지가 유독 많고, 유가가 쌌던 동북권에는 알뜰주유소 등이 많을까?

```python
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

#1. 서울을 서북, 서남, 동북, 동남으로 4분한다.

df6['구'].unique()

df6['서북권'] = df6['구'].isin(['은평구','서대문구','종로구','중구','용산구','마포구'])
df6['서남권'] = df6['구'].isin(['강서구','양천구','구로구','영등포구','금천구','관악구','동작구'])
df6['동남권'] = df6['구'].isin(['강동구', '강남구', '송파구', '서초구'])
df6['동북권'] = df6['구'].isin(['도봉구', '노원구', '강북구', '성북구', '중랑구', '동대문구', '성동구', '광진구'])
df6
```

```python
#2 - 4분한 것을 기반으로 하여 테이터를 소분하고, 파이차트를 통해 각 브랜드의 점유율을 구한다. 
df_sb = df6[df6['서북권']==True]
df_sb = df_sb.pivot_table(index='상표',aggfunc='sum').drop(axis=1,columns=['동남권','동북권','서남권','유가평균']).rename(columns={'서북권':'count'})
df_sn = df6[df6['서남권']==True]
df_sn = df_sn.pivot_table(index='상표',aggfunc='sum').drop(axis=1,columns=['동남권','동북권','서북권','유가평균']).rename(columns={'서남권':'count'})
df_db = df6[df6['동북권']==True]
df_db = df_db.pivot_table(index='상표',aggfunc='sum').drop(axis=1,columns=['동남권','서남권','서북권','유가평균']).rename(columns={'동북권':'count'})
df_dn = df6[df6['동남권']==True]
df_dn = df_dn.pivot_table(index='상표',aggfunc='sum').drop(axis=1,columns=['서남권','동북권','서북권','유가평균']).rename(columns={'동남권':'count'})

np.random.seed(134)
plt.figure(figsize=(30,40))
plt.rcParams['font.size'] = 22.0
plt.rcParams['figure.facecolor'] ='silver'
plt.rc('figure', titlesize=300) 


colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#eaff00','#ffc800'] #base color
wedgeprops={'width': 0.7, 'edgecolor': 'grey', 'linewidth': 5} #도넛모양 만들기. 

#pie1
plt.subplot(421)
colors1 = ['#ff9999', '#ffc000', '#8fd9b6', '#f5e1e7','#8fd2d9','#ffc800']
plt.pie(x='count',data=df_sb,
       autopct='%0.1f%%',startangle=0, counterclock=False,explode = np.random.rand(len(df_sb.index))*0.1
       ,colors=colors1[:len(df_sb.index)] , wedgeprops=wedgeprops)
plt.legend(['GS칼텍스', 'S-OIL', 'SK에너지', '자가상표', '현대오일뱅크'],bbox_to_anchor=(-0.2,1))
plt.title('서북권',fontsize=25,fontweight="bold")

#pie2
plt.subplot(422)
colors2 = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#8fd2d9','#ffc800']
plt.pie(x='count',data=df_sn,
       autopct='%0.1f%%',startangle=0, counterclock=False,explode = np.random.rand(len(df_sn.index))*0.1,
       colors=colors2[:len(df_sn.index)], wedgeprops=wedgeprops)
plt.legend(['GS칼텍스', 'S-OIL', 'SK에너지', '알뜰주유소', '현대오일뱅크'],bbox_to_anchor=(-0.2,1))
plt.title('서남권',fontsize=25,fontweight="bold")
#pie3
plt.subplot(423)
colors3 = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#f5e1e7','#8fd2d9']
plt.pie(x='count',data=df_db,
       autopct='%0.1f%%',startangle=0, counterclock=False,explode = np.random.rand(len(df_db.index))*0.1,
       colors=colors3[:len(df_db.index)], wedgeprops=wedgeprops)
plt.legend(['GS칼텍스', 'S-OIL', 'SK에너지', '알뜰주유소', '자가상표', '현대오일뱅크'],bbox_to_anchor=(-0.2,1))
plt.title('동북권',fontsize=25,fontweight="bold")

#pie4
plt.subplot(424)
colors4 = ['#ff9999',  '#517555','#ffc000','#8fd9b6','#a6d98f','#8fd2d9']

plt.pie(x='count',data=df_dn,
       autopct='%0.1f%%',startangle=0, counterclock=False,explode = np.random.rand(len(df_dn.index))*0.1,
       colors=colors4[:len(df_dn.index)], wedgeprops=wedgeprops)
plt.legend(['GS칼텍스', 'NH-OIL', 'S-OIL', 'SK에너지', '알뜰(ex)', '현대오일뱅크'],bbox_to_anchor=(-0.2,1))
plt.title('동남권',fontsize=25,fontweight="bold")
plt.show()
```

### 가정 4 결론. 
- 지역별로 브랜드 점유율이 다르다는 것은 대부분의 경우 주어진 데이터를 토대로 분석해본 결과 사실과 다르다. 
- 특정 브랜드의 가격이 유독 높다는 것 또한 SK에너지 혹은 GS칼텍스등의 대기업이 **지역과 상관없이 높은 점유율을 보유하고 있기 때문**에 나타난 결과로 보인다. 
- 따라서 브랜드의 지역화 또는 가격에 따른 브랜드의 밀집화는 사실과 일치하지 않는다.
