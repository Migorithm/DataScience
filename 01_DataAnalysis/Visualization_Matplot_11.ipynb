```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 딕셔너리를 이용한 DataFrame 생성
np.random.seed(100)
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
        'd':np.random.randn(50)
       }

#a - 0~49 
#c - 0~49사이 랜덤숫자  50개
#d - 정규분포 랜덤숫자 50개
data
```

```python
data['b'] = data['a']+10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
data
```

### DataFrame 시각화

```python
pd.DataFrame(data)
```

### 산점도 그래프 그리기 -> scatter
- 산점도 그래프는 scatter() 함수를 이용한다.
- X,Y 축에 해당하는 데이터의 상관관계... 얼마나 많이 분포 되어있고 분산은 어떠한가?
- X,Y 두개의 축을 기준으로 데이터가 얼마나 퍼져있는지를 알 수 있다. 

```python
#plt.scatter   (x,y,s,c) 
# x , y : coordinate
# s : 크기.
# c : 색깔. 직접 지정도 가능하고  c = range(4)  하는 식으로 4가지 랜덤 색깔을 넣을 수도 있다. 
#plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,350,500],c=['yellow','green','orange','red'])

plt.scatter('a','b',data=data,c='c',cmap='RdPu')
plt.colorbar()
plt.show()

#data['b'] 는 data['a']을 manipulate한 데이터기 때문에 양의 상관관계가 있는 건 당연하다. 
```

```python
plt.figure(figsize=(10,3))
plt.scatter('b','d',data=data,c=range(50),cmap='Greens')
plt.show()
```

```python
#plt.scatter   (x,y,s,c) 
# x , y : coordinate
# s : 크기.
# c : 색깔. 직접 지정도 가능하고  c = range(4)  하는 식으로 4가지 랜덤 색깔을 넣을 수도 있다. 


#plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,350,500],c=['yellow','green','orange','red'])
plt.scatter([1,2,3,4],[10,30,20,40],s=[100,200,350,500],c=range(4),cmap='Greens')
plt.colorbar()


plt.show()

#scatter 자체는 legend를 사용할 수 없다. 
```

```python
import seaborn as sns

#이상치값 제거. 
cutoff = (data['b']<10)
sns.set_style('darkgrid')
data['color'] = np.where(cutoff==True,'red','purple')
sns.regplot(x =data['a'],y=data['b'],scatter_kws={'facecolors':data['color']}) 

#scatter_kws : dictionaries
#Additional keyword arguments to pass to "plt.scatter" and "plt.plot".    
```

```python
import seaborn as sns

#이상치값 제거

cutoff = (data['a']>25) & (data['b'] >20)
cutoff

sns.set_style('dark') #전체적 톤을 줄 수 있다.

data['color'] = np.where(cutoff==True,"red","blue")
sns.regplot(x = data['a'], y=data['b'], scatter_kws={'facecolors':data['color']})
#컷 오프 조건에 맞으면 빨간색, 아니면 파란색이 될 것. 

plt.title("Scatter and Plot...", fontsize = 16)
plt.show()


#Seaborn provides an API on top of Matplotlib 
#that offers sane choices for plot style and color defaults, 
#defines simple high-level functions for common ...
```

### Subplot으로 plot(), scatter(), bar() 동시에 그려넣기.
- 영역을 쪼개서 여러개의 그래프를 표현하고 싶을 때 사용.

```python
x_values = ["A","B","C"]
y_values = np.array([1,10,100])

plt.figure(figsize=(12,4)) #전체 영역을 새롭게 잡아놓는다. 3등분하기 쉽도록.
plt.subplot(131)  #하나 지정하는데 세개 중 첫번째 -> 131
plt.plot(x_values,y_values)

plt.subplot(132)  #하나 지정하는데 세개 중 두번째 -> 132
plt.bar(x_values,y_values*2)

plt.subplot(133) #하나 지정하는데 세개 중 세번째 -> 133
plt.scatter(x_values,y_values*3 )

plt.show()

#통시에 같은 위치에서 다른 관점으로 접근하고자 할 때!
```
