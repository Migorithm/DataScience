### Matplot을 이용한 시각화

- 파이썬으로 데이터를 시각화할 때는 Matplotlib 라이브러리를 가장 많이 사용한다.
- Matplotlib은 파이썬에서 2D형태의 그래프, 이미지 등을 그릴 때 사용하는 것
- 과학 컴퓨팅 분야나 인공지능 분야에서도 많이 사용된다.

- Matplotlib 모듈에는 다양한 모듈들이 많이 있는데 그 중에서 가장 기본이 되는 pyplot이 서브 모듈.
    - import matplotlib.pyplot as plt
   


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#1. plot() :: 직선 혹은 꺾은 선의 그래프를 그릴 때 사용

#꺾은선 그래프.
plt.plot([1,2,3,4],[1,4,9,16]) # x축 , y 축 구조. 
plt.show()

#직선 그래프
plt.plot([10,20,30,40,50]) #인덱스값이 x축, 결과값이 y축에 간다.
plt.show()


# 그래프에 기본적인 옵션 추가하기. 
plt.plot([1,2,3,4],[1,4,9,16],'ro') #r = red , black... , 'ro' 
plt.axis([0,6,0,20]) # 앞의 두 인자가 x scale, 뒤의 두 인자가 y scale
plt.title("Graph Test")
plt.show()
```

### 겹쳐진 그래프

```python
import numpy as np
np.random.seed(100)
```

```python
#0부터 5미만까지 .... 0.2간격으로
t = np.arange(0,5,0.2)
t
plt.plot(t,t,'r--') #Dash line
plt.show()
```

```python
plt.plot(t,t,'r--') #Dash line
plt.plot(t,t**2,'bo-')
plt.plot(t,t**3,'g^') #green triangle
plt.show()
```

### 겹쳐진 그래프에 옵션 지정하기 

```python
plt.plot(t,t,'r--',label='A') #Dash line
plt.plot(t,t**2,'bo-',label='B')
plt.plot(t,t**3,'g^',label='C')

plt.legend(bbox_to_anchor=(1,1)) #붙인 Label들을 표현해주기 위해 필수 + bbox_to_anchor 통해서 위치 지정
plt.title("Overrapped Graph") #타이틀 설정
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.show()
```

### 파형을 나타내는 그래프 

```python
array=np.arange(0,13,0.001)
### 파형을 나타내는 그래프 

plt.figure(figsize=(10,6)) #프레임 크기는 가장 먼저 지정되어야. 
plt.plot(array,np.sin(array),'pink',label='sin graph',lw=3) #lw = line width. 선의 굵기 조절
plt.plot(array,np.cos(array),'skyblue',label='cos graph')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Wave Graph")
plt.show()
```
