```python
import numpy as np
import pandas as pd
from pandas import DataFrame as DF, Series as SS
import matplotlib.pyplot as plt
```

```python
tipdf = pd.read_csv('../data/tips.csv')
```

```python
tipdf.loc[245] = np.nan #you can't enlarge object with iloc as index cannot access non-existing row.
tipdf.loc[246] = np.nan 
tipdf
```

```python
#가장 팁이 많은 데이터 3개의 정보를 디스플레이
tipdf.sort_values(by=['tip'],ascending=False).head(3)
```

```python
def get_tippc(df): #인자값이 DataFrame
    return (df['tip']/df['total_bill'] *100).round(2) 

```

```python
tipdf['tip_pct'] = get_tippc(tipdf)
```

```python
#성별, 팁비율 별로 정렬, 높은거 3개만 추출.. 내림차순 정렬
tipdf.sort_values(['tip_pct'],ascending=[False]).groupby(by='sex').head(3).dropna()
```

```python
# Male 정보를 따로 분리해서 크기를 조회.
tipdf_man = tipdf[tipdf['sex']=='Male']

# or tipdf_man = tipdf[tipdf['sex'].isin(['Male'])] - the above one is going to be faster though.
tipdf_man.shape
```

```python
tipdf_woman = tipdf[tipdf['sex']=='Female']
tipdf_woman.shape
```

```python
#tipdf_female , tipdf_man을 단순하게 결합
tip_all = pd.concat([tipdf_man,tipdf_woman],keys=['Man','Woman'],axis=0)
```
