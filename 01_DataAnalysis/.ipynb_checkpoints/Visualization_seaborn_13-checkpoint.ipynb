### Seaborn의 plot visualization

### 1. paiplot
- 카테고리형 데이터(여러 품종의 붓꽃 데이터)가 섞여 있는 경우에는 hue 인수에 카테고리 이름을 지정한다.
- 품종에따라 색상을 다르게 하여 한꺼번에 비교, 분석할 수 있다.
- http://seaborn.pydata.org/generated/seaborn.pairplot.html


```python
sns.pairplot(iris) # 2**#columns 만큼 리턴.
```

```python
sns.pairplot(iris,hue='species')
```

```python
#일부만 보고 싶다면? 
sns.pairplot(iris,hue='species',vars=['sepal_length','sepal_width'])
```

```python
#x, y를 따로따로 지정 
sns.pairplot(iris,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
```

### 2. countplot
- http://seaborn.pydata.org/generated/seaborn.countplot.html
- 가장 많이 사용되는 시각화 툴 중 하나.
- 각 컬럼 별 데이터가 몇 개 있는지를 확인할 수 있다. 
- **해당 명령 함수는 DataFrame에만 적용할 수 있다.**
<br><br>
- Y축은 default로 x축에 해당하는 데이터의 갯수를 나타낸다. 
- 하지만 분석하려는 x에 실수값이나 또 다른 정수값이 들어간다면 Y에 해당 컬럼을 지정하고 Barplot을 사용하길 추천한다.

```python
titanic.columns
```

```python
titanic.sort_values(by='fare',ascending=False)
```

```python
titanic['class']
```

```python
#타이타닉 호의 생존자수 확인. 
sns.countplot(x='survived',data=titanic)

plt.title('# of Survivors from Titanic')
plt.show()
```

```python
#선실별 승객수
sns.countplot(x='class',data=titanic)
plt.title('# of passengers depending on class')
```

```python
#선실별 승객수 그리고 생존자. 
sns.countplot(x='class',data=titanic,hue='survived')
plt.title('# of passengers depending on class')
```

### 3. barplot

- countplot과 다르게 Y축 설정이 필요한 경우
- 숫자데이터로 Y축을 지정한다.
- 결과적으로 타이타닉호 등석에 따른 운임요금은 승객의 생존여부와 밀접한 관련이 있음을 알 수 있다.

- http://seaborn.pydata.org/generated/seaborn.barplot.html


```python
sns.barplot(x='class',y='fare',data=titanic,hue='survived')
```

### 4. pointplot

```python
sns.pointplot(x='class',y='fare',data=titanic,hue='survived')
plt.title('객실별 운임요금별 타이타닉호의 생존자 추이') #정말 이게 생존자 추이인가? 
sns.pointplot?
```
