### Histogram
- hist() 함수 사용
- 데이터의 빈도에 따라서 막대의 높이가 결정된다.
<br><br>
- 그래프들의 용도
    - plot()  : 꺾은 선 그래프
    - scatter()  : 산포도
    - hist()  : 빈도수

```python
plt.hist(np.random.randint(1,11,10)) 
#y값을 요구하지 않는 이유는 y값은 빈도수로 셋되기 때문이다.
```

```python
#bins option?

dice = []
for i in range(5):
    dice.append(np.random.randint(1,7))
dice
```

```python
'''
히스토그램에서는 bins값을 적당히 설정하는 것이 중요.
bins값이 너무 크면 막대가 너무 촘촘해서 이상하고
너무 작으면 막대가 뭉뚱그려져서 문제.
대개의 경우 bins값을 입력 안하면 기본값으로 잡히기 때문에 
특별한 경우가 아니라면 생략한다. 

bins
해당 막대의 영역을 얼마나 채울지를 결정하는 변수.
만약 bins가 6이라면 6+1을 범위로 나눈 값으로 너비를 입력한다.



'''
plt.hist(dice)
plt.show()
```

```python
#주사위를 백만번 돌렸을때 각 1에서 6까지의 숫자가 나오는 빈도수

'''
큰수의 법칙
던지는 횟수를 늘릴수록 특정한 숫자가 나오는 횟수가 기대값: P(E)에 가까워 진다는 수학적인 학설
'''

for i in range(1000000):
    dice.append(np.random.randint(1,7))
plt.hist(dice)
plt.show()
```

```python
mu , sigma = 100, 15   #mu= 평균 , sigma 표준편차 
x = mu + sigma*np.random.randn(1000000)

plt.hist(x,bins=250,color='g',density=1,alpha=0.3) #평균점이 mu(100)에 맞춰진 시그마값을 15로 갖는 정규분포를 만들겠다.
plt.grid()
plt.text(40,0.020,r"$\mu=100, \sigma=15$") #특수문자를 넣고 싶은 경우 
plt.show()

#plt.hist(x,bins=60,facecolor='g',alpha=0.3,density=1) 
#facecolor == color
#alpha = 선명도 
#density = 전체를 주어진 value로 보았을때 확률분포 (1을 주면 전체를 1(100%)로 보았을때의 범위를 의미하게된다. )
```

```python
plt.hist(x) # 세가지 값을 리턴한다. ->  n, bins, patches

n, bins, patches = plt.hist(x)

print(n)   # 빈도수
print(bins) # 구간 위치 
print(patches) 
```

```python
#값을 세밀하게 보고 싶다면.. bins로 조절
n, bins, patches = plt.hist(x,bins=60)

# len(bins) - > 61 
```

```python
#facecolor : 색깔 바꿀 때
#density=1 : 1을 기준으로 상대적 빈도확률
#alpha : 0~1사이의 실수값으로 투명도 조절

n, bins, patches = plt.hist(x,bins=60,facecolor='g',alpha=0.3,density=1)
plt.xlabel("Smarts")
plt.ylabel("Probability")
plt.title("Histogram of IQ")
plt.grid()

plt.text(60,0.025,r'$mu=100, \sigma=15$') 
#텍스트 기입하는 위치를 좌표로 정함. 따라서 그리드가 필요.
#r'' r 을 붙여서 raw문자열을 만들 수 있다. 이 문자열에서는 이스케이프 문자가 적용되지 않고 인용 부호 안의 문자가 그대로 출력됩니다.

#필요없는 부분 자르기
plt.axis([40,160,0,0.028])
plt.show()
```

### Box Plot


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#0부터 100까지 총 101개의 숫자를 만든다
xs = np.array(np.linspace(0,100,101))

df = pd.DataFrame(xs,columns=['feature'])
```

```python
#간단한 박스플롯을 만들어본다.
plt.figure(figsize=(7,6))
df.boxplot(column=['feature']) #column, NOT column's'
plt.yticks(np.arange(0,101,step=5),rotation=20) #y축 재 설정 
plt.show()
```

```python
tips = sns.load_dataset('tips')
#로컬저장소에 있는 tips가 아니라 anaconda lib에 있는 tips를 불러온다.

iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
titanic = sns.load_dataset('titanic')

flights

```

```python
#1. total_bill 값을 가지고 tip데이터의 중앙값과 50%안에 들어가는 tip값을 살펴보자.

plt.figure(figsize=(6,3))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
plt.show()
#밝을 수록 더 값이 큰 것!
#여름 철에 더 승객수가 더 많고
# 시간이 지날수록 더 많은 경향. 
```

```python
#값을 정확하게 알고 싶다면 annot 를 사용하자. 
#fmt 은 annotation의 포맷을 지정 d = > integer

plt.figure(figsize=(10,8))
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'), annot=True, fmt='d' )
plt.show()
```
